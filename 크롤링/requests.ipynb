### BeautifulSoup
 - 웹 문서 안에 있는 데이터를 추출할 수 있도록 함수를  제공해주는 모듈
 - 1. req 통해 html정보(텍슽) --> 파이썬 개체로 변환
 - 2. bs 제공하는 함수를 통해 통해 원하는 태그 정보에 접근하여 데이터 추출

```python
from bs4 import BeautifulSoup as bs
```

```python
# req 통해 html정보(텍슽) --> 파이썬 개체로 변환
html = bs(res.text,'lxml')# 파싱 parsing
html
```

```python
# css 선택자를 활용하여 접근할 수 있게 하는 함수
# html.select_one(css선택자): css선택자에 맞는 가장 먼저 찾아지는 첫번째 html 객체를 접근
# csss 선택자
# 태그선택자 : 태그명
# 아이디 선택자 : #아이디명
# 클랙스 선택자 :.클래스명
# 계층 선택자: 자식(>), 자손(공백),형제(~),인접형제(+)
html.select_one('title').text
```

```python
# html.select(css선택자): css 선택자에 맞는 html 객체 모두 접근
# 결과가 리스트 형식으로 반화
html.select('title')[0].text
```

```python
news_url = "https://n.news.naver.com/article/660/0000023947?cds=news_media_pc&type=editn"
news_url
```

```python
res = req.get(news_url)
res
#dpfj > 컴퓨터에게 브라우저에서 요청했다는 인식을 설정할 필요가 있음
```

```python
# 브라우저 요청 인식 설정
# 모든 페이지를 요청할 때 header값 지정할 것 ! 
header = {"user-agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.5304.122 Safari/537.36'}
```

```python
news_url = "https://n.news.naver.com/article/660/0000023947?cds=news_media_pc&type=editn"
res = req.get(news_url, headers = header)
res
```

```python
#파이썬 객체 변환
news_html = bs(res.text,'lxml')
news_html
```

```python
# 뉴스내용
a = news_html.select_one('#title_area > span').text.strip('\n')# strip(문자 양옆 공백제거) 왼쪽은 lstrip
b = news_html.select_one('#dic_area').text.strip('\n')
print(f'{a}\n{b}')
```
